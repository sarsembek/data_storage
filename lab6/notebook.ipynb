```python
class Node:
  def __init__(self, impurity=0, feature=-2, value=0):
    self.right = None
    self.left = None
    self.impurity = impurity
    self.feature = feature
    self.value = value

  def __repr__(self):
    return "leaf" if self.feature == -2 else feature_names[self.feature]
```

```python
values = clf.tree_.value.sum(axis=2).flatten()
nodes = [ Node(impurity, feature, value) \
 for impurity, feature, value in \
 zip(clf.tree_.impurity, clf.tree_.feature, values) ]
nodes
```

```python
def traverse_tree(nodes, i):
  if nodes[i].feature == -2:
    return 1

  nodes[i].right = nodes[i+1]
  offset_right = traverse_tree(nodes, i+1)
  nodes[i].left = nodes[i+offset_right+1]
  offset_left = traverse_tree(nodes, i+offset_right+1)
  return offset_right + offset_left + 1
traverse_tree(nodes, 0)
```

```python
def pre_order(node, visited):
  visited.append(node)
  if node.feature==-2:
    return
  pre_order(node.right, visited)
  pre_order(node.left, visited)
po_visit = []
pre_order(nodes[0], po_visit)
print(po_visit == nodes)
```

```python
def pre_order_id(node, tot_size, visited):
 left, right = node.left, node.right
 if node.feature == -2:
   visited.append(0)
   return
 decr = (node.impurity * node.value \
 - left.impurity * left.value \
 - right.impurity * right.value) / tot_size
 visited.append(decr)
 pre_order_id(node.right, tot_size, visited)
 pre_order_id(node.left, tot_size, visited)
impurity_decreases = []
pre_order_id(nodes[0], nodes[0].value, impurity_decreases)
impurity_decreases
```

```python
feature_importances = np.zeros(X.shape[1])
for feature, decr in zip(clf.tree_.feature, impurity_decreases):
 feature_importances[feature] += decr
feature_importances /= feature_importances.sum()
dict(zip(feature_names, feature_importances))
```

```python
def draw_split(features, thresholds, i, x0_min, x0_max, x1_min, x1_max):
    f, t = features[i], thresholds[i]
    if f == -2:
        # leaf node, return offset to be added
        # for the parent to visit the next child
        return 1
    # draw the current split
    if f == 0: # vertical split (x0)
        plt.axvline(t, x1_min * .1, x1_max * .1)
        # prepare the new bounding boxes
        # from x0_min to threshold
        bb_left = (x0_min, t, x1_min, x1_max)
        # from threshold to x0_max
        bb_right = (t, x0_max, x1_min, x1_max)
    else: # horizontal split (x1)
        plt.axhline(t, x0_min * .1, x0_max * .1)
        bb_left = (x0_min, x0_max, x1_min, t)
        bb_right = (x0_min, x0_max, t, x1_max)
    
    off_left = draw_split(features, thresholds, i + 1, *bb_left)
    off_right = draw_split(features, thresholds, i + off_left + 1, *bb_right)
    return off_right + off_left + 1

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlim(0, 10)
plt.ylim(0, 10)
draw_split(clf.tree_.feature, clf.tree_.threshold, 0, 0, 10, 0, 10)
```

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np

class MyRandomForestClassifier:
    def __init__(self, n_estimators=10, max_features='sqrt'):
        # Инициализируем список деревьев
        self.trees = [DecisionTreeClassifier(max_features=max_features) for _ in range(n_estimators)]

    def fit(self, X, y):
        for tree in self.trees:
            # Выбираем подмножество данных с заменой
            subset = np.random.choice(X.shape[0], size=X.shape[0], replace=True)
            # Используем .iloc для выборки по порядковым номерам, если X и y это DataFrame или Series
            X_subset = X.iloc[subset] if hasattr(X, 'iloc') else X[subset]
            y_subset = y.iloc[subset] if hasattr(y, 'iloc') else y[subset]
            # Обучаем дерево на подмножестве данных
            tree.fit(X_subset, y_subset)

    def predict(self, X):
        # Получаем предсказания всех деревьев
        predictions = np.array([tree.predict(X) for tree in self.trees])
        
        # Находим наиболее частое значение с помощью numpy
        def most_common(arr):
            values, counts = np.unique(arr, return_counts=True)
            return values[np.argmax(counts)]
        
        # Применяем most_common к каждому столбцу
        final_predictions = np.apply_along_axis(most_common, axis=0, arr=predictions)
        return final_predictions
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

class PseudoRandomForestClassifier:
    def __init__(self, n_estimators, max_features):
        self.trees = [DecisionTreeClassifier() for _ in range(n_estimators)]
        self.max_features = max_features
        self.features = []

    def fit(self, X, y):
        self.features = []
        for tree in self.trees:
            f_sub = np.random.choice(range(X.shape[1]), size=self.max_features, replace=False)
            subset = np.random.choice(range(X.shape[0]), size=X.shape[0], replace=True)
            # Используем .iloc для Pandas DataFrame
            X_subset = X.iloc[subset, f_sub] if hasattr(X, 'iloc') else X[subset][:, f_sub]
            y_subset = y.iloc[subset] if hasattr(y, 'iloc') else y[subset]
            tree.fit(X_subset, y_subset)
            self.features.append(f_sub)

    def predict(self, X):
        # Используем .iloc для выборки признаков в случае Pandas DataFrame
        predictions = [
            tree.predict(X.iloc[:, f_sub] if hasattr(X, 'iloc') else X[:, f_sub])
            for tree, f_sub in zip(self.trees, self.features)
        ]
        predictions = np.array(predictions, dtype=int)

        # Получаем наиболее частое предсказание
        def most_common(arr):
            values, counts = np.unique(arr, return_counts=True)
            return values[np.argmax(counts)]

        final_predictions = np.apply_along_axis(most_common, axis=0, arr=predictions)
        return final_predictions
```

```python
def rf_feature_importance(rf):
    fi_trees = np.array([tree.feature_importances_ for tree in rf.estimators_])
    return fi_trees.sum(axis=0) / fi_trees.sum()
```
